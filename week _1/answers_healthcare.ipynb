Convert column types to meaningful types 


1.   Convert one column to string
2.   Convert one column to float 
3.   Convert one column to bool 
4.   Convert one column to int
5.   Convert one column to datetime




<h1> 1 Convert one column to string


```python
aca_coverage['State'] = aca_coverage['State'].astype('string')
aca_coverage.dtypes
```

<h1> 2 Convert one column to float</h1> 


```python
aca_coverage['Marketplace Tax Credits (2016)'] = aca_coverage['Marketplace Tax Credits (2016)'].astype('float')
aca_coverage.dtypes
```

<h1> 3 Convert one column to bool


```python
aca_coverage['State Medicaid Expansion (2016)'] = aca_coverage['State Medicaid Expansion (2016)'].astype('bool')
aca_coverage.dtypes
```

<h1> 4 Convert one column to int


```python
# Explains the code below:
# aca_coverage['Average Monthly Tax Credit (2016)'] is us calling the column we are wanting the code to work in. 
# aca_coverage['Average Monthly Tax Credit (2016)'] = aca_coverage['A... is writing over the column so the values stay in place when called later. 
# .str.replace(' ', ''), it replaces all occurrences of a space with an empty string, effectively removing the spaces from each element in the DataFrame column.
# .str.lstrip("$") left strips the $ as its to the left of the object
# finally the .astype('int') converts the object to an int now that it meets the Required parameters

aca_coverage['Average Monthly Tax Credit (2016)'] = aca_coverage['Average Monthly Tax Credit (2016)'].str.replace(' ', '').str.lstrip("$").astype('int')
aca_coverage.info()
```

<h1> 5 Convert one column to datetime

```python
Surescripts_Network['Period'] = pd.to_datetime(Surescripts_Network['Period'])
Surescripts_Network.head()
```
